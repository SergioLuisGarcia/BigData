```python
import pandas as pd
import matplotlib.pyplot as plt
df1 = pd.read_csv("2020parte1.csv", sep=";")
df2 = pd.read_csv("2020parte2.csv", sep=";")
df3 = pd.read_csv("2021parte1.csv", sep=";")
df4 = pd.read_csv("2021parte2.csv", sep=";")
df5 = pd.read_csv("2022parte1.csv", sep=";")
df6 = pd.read_csv("2022parte2.csv", sep=";")
df_2020 = pd.merge(df1,df2,how='outer')
df_2021 = pd.merge(df3,df4,how='outer')
df_2022 = pd.merge(df5,df6,how='outer')
```

```python
df_2020 = df_2020.dropna(subset=['estado','municipio'])
df_2021 = df_2021.dropna(subset=['estado','municipio'])
df_2022 = df_2022.dropna(subset=['estado','municipio'])
```

```python
dados_sergipe_2020 = df_2020.query('estado=="SE"')
capital_2020 = dados_sergipe_2020.query('municipio == "Aracaju"')
print(f"Total de casos em 2020 na capital {capital_2020['casosNovos'].sum()}")
```

```python
dados_sergipe_2021 = df_2021.query('estado=="SE"')
capital_2021 = dados_sergipe_2021.query('municipio == "Aracaju"')
print(f"Total de casos em 2021 na capital {capital_2021['casosNovos'].sum()}")
```

```python
dados_sergipe_2022 = df_2022.query('estado=="SE"')
capital_2022 = dados_sergipe_2022.query('municipio == "Aracaju"')
print(f"Total de casos em 2022 na capital {capital_2022['casosNovos'].sum()}")
```

```python
anos = [2020,2021,2022]
casos_capital = [capital_2020['casosNovos'].sum(),
                 capital_2021['casosNovos'].sum(),
                 capital_2022['casosNovos'].sum()]

plt.figure(figsize=(25, 10))

plt.subplot(131)
plt.bar(anos, casos_capital)
plt.suptitle('Grafico de casos por ano na capital:')
plt.show()
print(f"Total de casos:{capital_2020['casosNovos'].sum() +capital_2021['casosNovos'].sum() +capital_2022['casosNovos'].sum()}")
```

```python
geral_2020 = dados_sergipe_2020['casosNovos'].sum()
print(f"Total de casos no estado de Sergipe em 2020:{geral_2020}")
```

```python
geral_2021 = dados_sergipe_2021['casosNovos'].sum()
print(f"Total de casos no estado de Sergipe em 2021:{geral_2021}")
```

```python
geral_2022 = dados_sergipe_2022['casosNovos'].sum()
print(f"Total de casos no estado de Sergipe em 2022:{geral_2022}")
```

```python
anos = [2020,2021,2022]
casos_gerais = [geral_2020,geral_2021,geral_2022]

plt.figure(figsize=(25, 10))

plt.subplot(131)
plt.bar(anos, casos_gerais)
plt.suptitle('Grafico de casos por ano no estado:')
plt.show()
print(f"Total de casos:{geral_2020+geral_2021+geral_2022}")
```
